```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
import matplotlib.font_manager as font_manager

N = 100
x = np.linspace(0.0, 1.0, N)

y = 4 * x + np.sin(x) - 2 * np.sin(2 * x) - 4 * np.sin(10 * x)

z = np.angle(np.exp(1.0j * y))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)

ax.plot(x, y, color='red', label="original function f(x)")
ax.plot(x, z, color='blue', label="arg{exp(if(x))}")
ax.plot(x, np.pi * np.ones(N), color='gray', label='pi')
ax.plot(x, -np.pi * np.ones(N), color='gray', label='-pi')

ax.legend()
plt.show()

fig.savefig('../../thesis/chap2/figure/before_unwrap.png')
```

```python
prev = z[0]
w = z.copy()
for i in range(N):
    if z[i] < prev - math.pi:
        w[i] = z[i] + 2.0 * math.pi
    elif z[i] > prev + math.pi:
        w[i] = z[i] - 2.0 * math.pi
    else:
        w[i] = z[i]
    prev = w[i]
```
