```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def wave_func(r):
    return r - 1.05*r**2 + 0.25*(3*r**3 - 2*r)


N = 100
r = np.linspace(0, 1.0, N)
y = wave_func(r)

plt.plot(r, y)
```

```python
def draw(N):
    fig = plt.figure(figsize=(10, 8))
    r = np.linspace(0, 1.0, N)
    y = wave_func(r)

    plt.title('resolution: ' + str(N) + ' px', fontsize=24)
    plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
    plt.axis('off')
    plt.plot(r, y, linewidth=10)

    fig.savefig('../../thesis/chap1/figure/radial_resolution_' + str(N) + '.png')
```

```python
draw(3)
```

```python
draw(5)
```

```python
draw(10)
```

```python
draw(100)
```
