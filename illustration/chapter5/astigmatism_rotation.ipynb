```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
before = np.fromfile('../../data/astigmatism_rotation/202101092101.dat', dtype=np.uint16).astype(np.float)
after = np.fromfile('../../data/astigmatism_rotation/202101092149.dat', dtype=np.uint16).astype(np.float)
```

```python
N = int(math.sqrt(before.shape[0]))
before = before.reshape((N, N))
```

```python
M = int(math.sqrt(after.shape[0]))
after = after.reshape((M, M))
```

```python
center_x = 1010
center_y = 880
K = 128
before_center = before[(center_x-K//2):(center_x+K//2), (center_y-K//2):(center_y+K//2)]

fig = plt.figure(figsize=(8, 8))
plt.axis('off')
plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
plt.imshow(before_center)
fig.savefig('../../thesis/chap5/figure/before_rotation.png')
```

```python
K2 = 128

center_x = M//2-5
center_y = M//2+3

after_center = after[(center_x-K2//2):(center_x+K2//2), (center_y-K2//2):(center_y+K2//2)]

fig = plt.figure(figsize=(8, 8))
plt.axis('off')
plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
plt.imshow(after_center)
fig.savefig('../../thesis/chap5/figure/after_rotation.png')
```
