```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
N = 1000
x = np.linspace(-1.0, 1.0, N)
y = x.copy()
x, y = np.meshgrid(x, y)
r = np.hypot(x, y)
theta = np.arctan2(y, x)

on_ring = (0.5 <= r) & (r <= 0.7)
dist = np.zeros((N, N))
dist = r**2 * np.cos(2*theta) * on_ring

fig = plt.figure(figsize=(8, 8))

plt.xlabel('x [mm]', fontsize=18)
plt.ylabel('y [mm]', fontsize=18)
plt.xticks([0, 200, 400, 600, 800, 999], 2.0 * (np.array([0, 200, 400, 600, 800, 999], dtype=np.float32) / N - 0.5))
plt.yticks([0, 200, 400, 600, 800, 999], 2.0 * (np.array([0, 200, 400, 600, 800, 999], dtype=np.float32) / N - 0.5))
plt.imshow(dist)
fig.savefig('../../thesis/chap3/figure/unwrap_xy.png')
```

```python
r = np.linspace(0.5, 0.7, 100)
theta = np.linspace(0.0, 2.0 * math.pi, 1000)
theta, r = np.meshgrid(theta, r)

on_ring = (0.5 <= r) & (r <= 0.7)
dist_polar = r**2 * np.cos(2 * theta) * on_ring

fig = plt.figure(figsize=(8, 6))
plt.xlabel('theta [rad]', fontsize=18)
plt.ylabel('r [mm]', fontsize=18)
plt.xticks([0, 200, 400, 600, 800, 999], 2.0 * math.pi * np.array([0, 200, 400, 600, 800, 999], dtype=np.float32) / 1000)
plt.imshow(dist_polar)
fig.savefig('../../thesis/chap3/figure/unwrap_polar.png')
```
