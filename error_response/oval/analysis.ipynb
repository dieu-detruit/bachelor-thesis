```python
import numpy as np
import matplotlib.pyplot as plt
import math
from PIL import Image
```

```python
peak_data = np.loadtxt('xray_strehl/peak.txt').transpose()
delta, peak = peak_data
```

```python
strehl = peak / np.max(peak)

fig = plt.figure(figsize=(10, 5))

plt.xlabel('$\delta_{\mathrm{pv}}$ [nm]', fontsize=24)
plt.ylabel('Strehl ratio [/]', fontsize=24)
plt.xticks(fontsize=16)
plt.yticks(fontsize=24)

plt.plot(1.0e+9 * delta, strehl)

fig.savefig('../../thesis/chap2/figure/error_response/oval/strehl.png', bbox_inches="tight")
```

```python
diff = np.abs(strehl - 0.8)
print('allowed error:', delta[np.argmin(diff)])
print('index:', np.argmin(diff))
```

```python
strehl_just_focus = np.array(Image.open('xray_strehl/just/focus_abs.png'))

info_file = open('xray_strehl/just/focus.info')
info = info_file.read().split()
print(info)
info_file.close()

N = int(info[0])
K = 128
x = np.linspace(float(info[2]), float(info[3]), N)
x = np.round(1.0e+9 * np.linspace(x[N//2-K//2], x[N//2+K//2-1], 7))

print(N)
print(x)

fig = plt.figure(figsize=(8, 8))

plt.imshow(strehl_just_focus[(N//2-K//2):(N//2+K//2), (N//2-K//2):(N//2+K//2)], origin='lower')
plt.xticks(np.linspace(0, K-1, 7), x, fontsize=20)
plt.yticks(np.linspace(0, K-1, 7), x, fontsize=20)

plt.xlabel('x [nm]', fontsize=24)
plt.ylabel('y [nm]', fontsize=24)

fig.savefig('../../thesis/chap2/figure/error_response/oval/just_strehl.png')
```

```python
hpd_data = np.loadtxt('xray_hpd/hpd.txt').transpose()
x, hpd = hpd_data
```

```python
#hpd_arcsecond = hpd / 2.1025 / math.pi * 180 * 3600
hpd_arcsecond = hpd

fig = plt.figure(figsize=(10, 5))

plt.xlabel('$\delta_{\mathrm{pv}}$ [um]', fontsize=24)
plt.ylabel('HPD [arcsecond]', fontsize=24)
plt.xticks(fontsize=16)
plt.yticks(fontsize=24)

plt.plot(1.0e+6 * x, hpd_arcsecond)

fig.savefig('../../thesis/chap2/figure/error_response/oval/hpd.png', bbox_inches="tight")
```

```python
diff = np.abs(hpd_arcsecond - 0.5)
print('allowed error:', x[np.argmin(diff)])
print('index:', np.argmin(diff))
```
